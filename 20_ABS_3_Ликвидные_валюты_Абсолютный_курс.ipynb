```
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt

import yfinance as yf
```

# Готовим список пар и валют

```
# https://admiralmarkets.com/ru/education/articles/forex-basics/osnovnye-valiutnye-pary-foreks
curParStr = 'AUD/USD EUR/USD GBP/USD USD/CAD USD/CHF USD/JPY'

curPars = np.unique(curParStr.replace('/','').split())
print('валютные пары',curPars)

curs = np.unique(curParStr.replace('/',' ').split())
print('валюты',curs)
```

```
data = yf.download([x+'=x' for x in curPars])
data
```

# Отбираем курсы закрытия

```
close = data.Close.dropna()
close
```

# Графики пар

```
for cur in close.columns:
    print(cur)
    plt.figure(figsize=(12,7))
    close[cur].plot()
    plt.grid()
    plt.title(cur)
    plt.savefig(cur+'.jpg')
    plt.show()
    
```

# Относительные изменения курса

```
for cur in close.columns[0:6]:
    d_close = close[cur].pct_change()
    plt.figure(figsize=(12,7))
    d_close.plot()
    plt.grid()
    plt.title("d_" + cur)
    plt.savefig("d_" + cur+'.jpg')
    plt.show()
    print(d_close.describe())
```
