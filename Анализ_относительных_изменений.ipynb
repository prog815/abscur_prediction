Есть таблица с выгрузками абсолютных курсов. Берем ее данные и в колабе анализируем. Сначала импортируем. Затем рисуем графики. Затем смотрим на статистические характеристики. 

# Библиотеки

```
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
```

# Импорт данных

Таблица с историей абсолютных курсов: https://docs.google.com/spreadsheets/d/1_-cdNCIC6NgyRDj0zLJBaohXo-K8qCfxfDLbYq2JLhs/edit#gid=422799220

```
data_url = 'https://docs.google.com/spreadsheets/d/1_-cdNCIC6NgyRDj0zLJBaohXo-K8qCfxfDLbYq2JLhs/export?format=csv&id=1_-cdNCIC6NgyRDj0zLJBaohXo-K8qCfxfDLbYq2JLhs&gid=422799220'
abs_data = pd.read_csv(data_url,decimal=',',parse_dates=True,index_col=0)
abs_data.shape
```

## Подготовка данных

```
dates = abs_data.index.values # даты
cur_names = abs_data.columns.values # имена валют
data = abs_data.values # матрица абсолютных курсов
d_data = np.zeros_like(data) # матрица относительных изменений
d_data[1:,:] = (data[1:,:]-data[:-1,:])/data[:-1,:] # расчет матрицы относительных изменений
```

# Вывод графиков (несколько примеров)

```
for n in np.random.randint(len(cur_names),size=6):
    plt.plot(dates,d_data[:,n])
    plt.grid()
    plt.title("Относительные изменения курса " + cur_names[n].upper())
    plt.xticks(rotation=20)
    plt.show()
```

# Статистические характеристики

```
d_data_stat = pd.DataFrame(d_data,index=dates,columns=cur_names).describe().T
d_data_stat
```

## Средние значения

```
col = d_data_stat['mean'].sort_values()
plt.figure(figsize=(10,7))
plt.title('Среднее')
cont = plt.bar(col.index,col.values)
plt.xticks(rotation=90)
plt.show()
```

Можно видеть кто в падении и кто в росте. Но на всем временном промежутке.

## Стандартные отклонения

```
col = d_data_stat['std'].sort_values()
plt.figure(figsize=(10,7))
plt.title('Стандартное отклонение')
cont = plt.bar(col.index,col.values)
plt.xticks(rotation=90)
plt.show()
```

Можно видеть кто более подвижен (изменчив) и кто менее менее.
